```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./ml_source_files/USA_Housing.csv')
```

```python
df.head(5)
```

```python
# 집값을 예측하는 모델
```

```python
y = df['Price']
```

```python
x = df.drop(['Price', 'Address'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
preds = lr.predict(x_test)
preds
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, preds)
```

```python
df = pd.read_csv('./ml_source_files/titanic_train.csv')
df.head(3)
```

```python
df.Survived.value_counts().plot(kind = 'bar')
```

```python
sns.countplot(df['Survived'], hue = 'Sex', data = df)
```

```python
sns.displot(df['Age'], bins = 50)
```

```python
sns.lmplot(x = 'Age', y = 'Survived', data = df, hue = 'Sex')
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin', axis = 1, inplace = True)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
y = df['Survived']
```

```python
x = df.drop(['Name', 'Ticket', 'Fare'], axis = 1)
```

```python
x = pd.get_dummies(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
preds = log.predict(x_test)
preds
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, preds)
```

```python
print(classification_report(y_test, preds))
```

```python
df = pd.read_csv('./ml_source_files/advertising.csv')
df.head(3)
```

```python
df.info()
```

```python
# 광고를 클릭할지 여부를 판별하는 모델을 생성
```

```python
y = df['Clicked on Ad']
```

```python
x = df[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 111)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
preds = log.predict(x_test)
```

```python
preds
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, preds)
```

```python
print(classification_report(y_test, preds))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./ml_source_files/train_santander.csv')
df.head()
```

```python
# Target => 불만족 : 1, 만족 : 0
```

```python
df.TARGET.value_counts()
```

```python
df.describe()
```

```python
# 이상치(outlier) 제거

df.var3.value_counts().sort_index()
df['var3'].replace(-999999, 2, inplace = True)
```

```python
# x, y data divide

y = df['TARGET']
x = df.drop('TARGET', axis = 1)
```

```python
x.drop('ID', axis = 1, inplace = True)
```

```python
# train test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=111)
```

```python
# algorithm import

from xgboost import XGBClassifier
```

```python
xgb_clf = XGBClassifier(n_estimators = 500, random_state = 111)
xgb_clf.fit(x_train, y_train, early_stopping_rounds = 100, eval_metric = 'auc',\
            eval_set = [(x_train, y_train), (x_test, y_test)])
```

```python
# model evaluation
```

```python
from sklearn.metrics import classification_report, roc_auc_score
```

```python
preds = xgb_clf.predict(x_test)
```

```python
print(classification_report(y_test, preds))
```

```python
roc_auc_score(y_test, xgb_clf.predict_proba(x_test)[:,1], average = 'macro')
```

```python
from sklearn.model_selection import GridSearchCV

xgb_cls = XGBClassifier(n_estimators = 100)

params = {'max_depth' : [5, 7], 'min_child_weight' : [1, 3],\
        'colsample_bytree' : [0.5, 0.7]}

grid = GridSearchCV(xgb_cls, param_grid = params, verbose = False)
grid.fit(x_train, y_train, early_stopping_rounds = 30, eval_metric = 'auc', \
        eval_set = [(x_train, y_train), (x_test, y_test)])
```

```python
xg_cls = XGBClassifier(n_estimators = 100, closample_bytree = 0.5, \
                       max_depth = 5, min_child_weight = 1)

xgb_cls.fit(x_train, y_train)
preds = xgb_cls.predict(x_test)                    
```
